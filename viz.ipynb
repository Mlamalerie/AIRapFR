```python
import pandas as pd
from glob import glob
import os,json
import matplotlib.pyplot as plt
```

```python
BACKUP_FOLDER_PATH = f"{os.getcwd()}\data"
BACKUP_FOLDER_PATH
```

```python
glob(f"data/*/*.csv")
```

```python
def get_all_csv_files(dir_path : str):
    if files := glob(f"{dir_path}/*/*.csv"):
        return pd.concat([pd.read_csv(file_path) for file_path in files],ignore_index=True)
    return None

df_all_data = get_all_csv_files("data")
df_all_data
```

```python
df_all_data.info(verbose=True)
```

```python
df_all_data["artist"].unique()
```

```python
df_all_data["lyrics"] = df_all_data["lyrics"].astype(str)
```

```python
sorted(list(df_all_data.columns))
```

```python

df_all_data["lyrics_words"] = df_all_data['lyrics'].apply(lambda x: str(x).split(" ") )
df_all_data["lyrics_words_count"] = df_all_data['lyrics'].apply(lambda x: len(x))
df_all_data["lyrics_unique_words_count"] = df_all_data['lyrics'].apply(lambda x: len(set(x)))

plt.figure(figsize=(11,6))
ax = plt.subplot(1,2,1)
ax.set_title("Repartition du nombre de mots par chansons")
df_all_data["lyrics_words_count"].plot.hist(bins=20)

ax = plt.subplot(1,2,2)
ax.set_title("Repartition du nombre de mots uniques par chansons")
df_all_data["lyrics_unique_words_count"].plot.hist(bins=20)

plt.show()
```

```python
df_all_data[["lyrics_words_count","lyrics_unique_words_count"]].describe()
```

```python

plt.figure(figsize=(11,45))
ax = plt.subplot(2,2,1)
ax.set_title("Nombre de lyrics")
df_all_data.artist.value_counts().sort_values().plot(kind='barh')

ax = plt.subplot(2,2,2)
ax.set_title("Nombre moyen de mots par lyrics")
df_all_data.groupby("artist").mean()["lyrics_words_count"].sort_values().plot(kind="barh")

ax = plt.subplot(2,2,3)
ax.set_title("Nombre moyen de mots unique par lyrics")
df_all_data.groupby("artist").mean()["lyrics_unique_words_count"].sort_values().plot(kind="barh")

plt.show()
```
