```python
import re
import requests
from bs4 import BeautifulSoup as bs
```

```python


def get_soup(url):
    r = requests.get(url)
    soup = bs(r.text, 'html.parser')
    return soup



def get_insultes_from_page_wiki(url):
    soup = get_soup(url)

    div = soup.find('div', {'id': 'mw-pages'})
    insultes = []
    for li in div.find_all('li'):
        insultes.append(li.text.lower())

    return set(insultes)

def get_insultes():
    urls = ["https://fr.wiktionary.org/w/index.php?title=Cat%C3%A9gorie:Insultes_en_fran%C3%A7ais&pageuntil=gland#mw-pages", "https://fr.wiktionary.org/w/index.php?title=Cat%C3%A9gorie:Insultes_en_fran%C3%A7ais&pagefrom=gland#mw-pages", "https://fr.wiktionary.org/w/index.php?title=Cat%C3%A9gorie:Insultes_en_fran%C3%A7ais&pagefrom=sac+%C3%A0+foutre#mw-pages"]

    result = set()
    for url in urls:
        result = result.union(get_insultes_from_page_wiki(url))
    return list(result)

insultes_fr = get_insultes()
insultes_fr
```

```python
def to_censored_word(word):
    return re.sub(r'[aieou]', '*', word)

censored_words_dict = {insulte : to_censored_word(insulte) for insulte in insultes_fr}
censored_words_dict
```
